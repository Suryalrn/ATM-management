```python
import json
```

```python
class ATM:
    def __init__(self):
        self.deposited_amount = 0
        self.withdrawn_amount = 0
        self.balance = 0
        self.card_number_inserted = 0
        self.card_number = 0
        self.pin = ""
        self.menu()
        
        

    def menu(self):
        print("Insert your ATM card to start the Process")
        self.card_inserted = int(input("Card insert"))
        if(self.card_inserted):
            self.card_number_inserted = int(input("Enter your card number: "))
            if(self.check_card_existence()):
                self.Process_start()
            else:
                with(open("data.json","r")as file):
                    data = json.load(file)
                self.card_number += 1
                data["cards"].append({"card_number":self.card_number,"balance":0,"pin":""})
                with(open("data.json","w")as file):
                    json.dump(data,file,indent=4)
                print(f"New card created. Your card number is {self.card_number}")
                self.Process_start()
        elif(self.card_inserted <= 0 or self.card_inserted == None):
            self.menu()
    
    def check_card_existence(self):
        with(open("data.json", "r")as file):#file = open("data.json", "r") file.close()
            data = json.load(file)
            for card in data["cards"]:
                if card["card_number"] == self.card_number_inserted:
                    return True
            return False
        
    def Process_start(self):
        if(self.pin ==""):
            self.pin = input("Create a pin number of 4 digits: ")
            if len(self.pin) == 4 and self.pin.isdigit():
                with(open("data.json", 'r')as file):
                    data = json.load(file)
                for card in data["cards"]:
                    if card[{"card_number"}] == self.card_number:
                        card["pin"] = self.pin
                with(open(data.json, 'w')as file):
                    json.dump(data, file, indent=4)
                print("Pin created successfully.")
            else:
                print("Invalid pin. Please try again.")
                self.Process_start()

        user_input = input("""
            press no. to make selection:
            1. Deposit
            2. Withdraw
            3. Check Balance
            4. Transfer
            5. Change pin 
            6. Exit
            """)
        match user_input:
            case "1":
                self.deposit()
            case "2":
                self.withdraw()
            case "3":
                self.check_balance()
            case "4":
                self.transfer()
            case "5":
                print("Thank you for using our services.")
                exit()
            case _:
                print("Invalid selection. Please try again.")
                self.menu()

    def deposit(self):
        amount = float(input("Enter the amount to deposit: "))
        if amount > 0:
            self.deposited_amount += amount
            self.balance += amount
            print(f"Deposited: ${amount:.2f}")
        else:
            print("Invalid deposit amount. Please try again.")
        self.menu()
        
    def withdraw(self):
        amount = float(input("Enter the amount to withdraw: "))
        if 0 < amount <= self.balance:
            self.withdrawn_amount += amount
            self.balance -= amount
            print(f"Withdrew: ${amount:.2f}")
        else:
            print("Invalid withdrawal amount or insufficient funds. Please try again.")
        self.menu()

    def check_balance(self):
        print(f"Current Balance: ${self.balance:.2f}")
        self.menu()
    
    def transfer(self):
        amount = float(input("Enter the amount to transfer:"))
        if amount > 0 and amount <= self.balance:
            self.balance -=amount;
            print(f"Transferred: ${amount:.2f}")
        else:
            print("Invalid transfer amount or insufficient funds. Please try again.")
        self.menu()

    
```
