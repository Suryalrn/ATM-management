```python
import json
```

```python
class ATM:
    def __init__(self):
        self.deposited_amount = 0
        self.withdrawn_amount = 0
        self.balance = 0
        self.card_number_inserted = 0
        self.card_number = 0
        self.card_exist = False
        self.menu()

    def load_data(self):
        with(open("data.json","r")as file):
            return json.load(file)

    def save_data(self,data):
        with(open("data.json","w")as file):
            json.dump(data,file,indent = 4)
        

    def menu(self):
        print("Insert your ATM card to start the Process")
        self.card_inserted = int(input("Card insert"))
        if(self.card_inserted):
            self.card_number_inserted = int(input("Enter your card number: "))
            if(self.check_card_existence()):
                self.Process_start()
            else:
                print("invalid no.")
                new_card = input("Do you want to create new card y/n??").lower()
                if new_card == "y":
                    self.create_card()
                    self.create_pin()
                    self.Process_start()
                else:
                    exit()  
        elif(self.card_inserted <= 0):
            self.menu()


    # Card_Work    
    
    def check_card_existence(self):
        self.data = self.load_data()
        if str(self.card_number_inserted) in self.data["cards"]:
                self.card = self.data["cards"][str(self.card_number_inserted)]
                return True
        return False
        
    def create_card(self):
        self.data = self.load_data()
        new_card = max(map(int, self.data["cards"]))+1
        self.data["cards"][str(new_card)]= {"balance":0,"pin":0}
        self.save_data(self.data)
        self.card = self.data["cards"][str(new_card)]
        print(f"New card created. Your card number is {new_card}")
        # Card End
        
    # Pin_Working 
    def create_pin(self):
        if self.card["pin"] == 0 :
            self.pin = int(input("Create a pin number of 4 digits: "))
            if(self.save_pin()):
                print("Pin created successfully.")
            else:
                print("Wrong format. Please try again.")
                self.create_pin()
        else:
            print("Pin already exists. Please try again.")
            self.Process_start()

    def save_pin(self):
        if 1000<=self.pin<=9999:
            self.card["pin"] = self.pin
            self.save_data(self.data)
            return True
        else:
            return False
        
    def pin_authentication(self):
        correct = False
        self.entered_pin = int(input("Enter your pin number: "))
        if(self.card["pin"] == self.entered_pin):
            correct = True
            return correct
        else:
            return correct
        
    def change_pin(self):
        if(self.pin_authentication()):
            new_pin = int(input("enter new pin"))
            self.card["pin"] = new_pin
            self.save_data(self.data)
            print("pin changed successfully")
            # Pin End
            
                                        
            
    def Process_start(self):
        if(self.pin_authentication()):
            print("Authentication Successful")
        else:
            print("wrong pin")
            self.Process_start()
        

        user_input = input("""
            press no. to make selection:
            1. Deposit
            2. Withdraw
            3. Check Balance
            4. Transfer
            5. Change pin 
            6. Exit
            """)
        match user_input:
            case "1":
                self.deposit()
            case "2":
                self.withdraw()
            case "3":
                self.check_balance()
            case "4":
                self.transfer()
            case "5":
                self.change_pin()
            case "6":
                print("Thank you for using our services.")
                exit()
            case _:
                print("Invalid selection. Please try again.")
                self.menu()

    def deposit(self):
        amount = float(input("Enter the amount to deposit: "))
        if amount > 0:
            self.card["balance"] += amount
            self.save_data(self.data)
            print(f"Deposited: ${amount:.2f}")
        else:
            print("Invalid deposit amount. Please try again.")
        self.menu()
        
    def withdraw(self):
        amount = float(input("Enter the amount to withdraw: "))
        if 0 < amount <= self.card["balance"]:
            self.card["balance"] -= amount
            self.save_data(self.data)
            print(f"Withdrew: ${amount:.2f}")
        else:
            print("Invalid withdrawal amount or insufficient funds. Please try again.")
        self.menu()

    def check_balance(self):
        self.balance = self.card["balance"]
        print(f"Current Balance: ${self.balance:.2f}")
        self.menu()
    
    def transfer(self):
        amount = float(input("Enter the amount to transfer:"))
        if amount > 0 and amount <= self.card["balance"]:
            self.card["balance"] -=amount
            self.save_data(self.data)
            print(f"Transferred: ${amount:.2f}")
        else:
            print("Invalid transfer amount or insufficient funds. Please try again.")
        self.menu() 

```
