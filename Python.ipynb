```python
import json
```

```python
class ATM:
    def __init__(self):
        self.deposited_amount = 0
        self.withdrawn_amount = 0
        self.balance = 0
        self.card_number_inserted = 0
        self.card_number = 0
        self.pin = 0000
        self.menu()
        

    def menu(self):
        print("Insert your ATM card to start the Process")
        self.card_inserted = int(input("Card insert"))
        if(self.card_inserted):
            self.card_number_inserted = int(input("Enter your card number: "))
            if(self.check_card_existence()):
                self.Process_start()
            else:
                self.create_card()
                self.create_pin()
                self.Process_start()
        elif(self.card_inserted <= 0 or self.card_inserted == None):
            self.menu()
    
    def check_card_existence(self):
        with(open("data.json", "r")as file):#file = open("data.json", "r") file.close()
            data = json.load(file)
            for card in data["cards"]:
                if card["card_number"] == self.card_number_inserted:
                    return True
            return False
        
    def create_card(self):
        with(open("data.json","r")as file):
            data = json.load(file)
        self.card_number = data["cards"][-1]["card_number"]+1
        data["cards"].append({"card_number":self.card_number,"balance":0,"pin":0000})
        with(open("data.json","w")as file):
            json.dump(data,file,indent=4)
        print(f"New card created. Your card number is {self.card_number}")
        

    def create_pin(self):
        if(self.pin ==0):
            self.pin = int(input("Create a pin number of 4 digits: "))
            if(self.save_pin()):
                print("Pin created successfully.")
            else:
                print("Wrong format. Please try again.")
                self.create_pin()
        else:
            print("Pin already exists. Please try again.")
            self.Process_start()

        
            
    def pin_authentication(self):
        correct = False
        self.entered_pin = int(input("Enter your pin number: "))
        with(open("data.json", 'r')as file):
            data = json.load(file)
            for card in data["cards"]:
                if card["card_number"] == self.card_number_inserted and card["pin"] == self.entered_pin:
                    correct = True
            return correct
    
    def read(self):
        with(open("data.json", 'r')as file):
            data = json.load(file)
            
                                        
            
    def Process_start(self):
        if(self.pin_authentication()):
            print("Authentication Successful")
        else:
            print("wrong pin")
            self.Process_start()
        

        user_input = input("""
            press no. to make selection:
            1. Deposit
            2. Withdraw
            3. Check Balance
            4. Transfer
            5. Change pin 
            6. Exit
            """)
        match user_input:
            case "1":
                self.deposit()
            case "2":
                self.withdraw()
            case "3":
                self.check_balance()
            case "4":
                self.transfer()
            case "5":
                self.change_pin()
            case "6":
                print("Thank you for using our services.")
                exit()
            case _:
                print("Invalid selection. Please try again.")
                self.menu()

    def deposit(self):
        amount = float(input("Enter the amount to deposit: "))
        if amount > 0:
            self.read()
            with(open("data.json", 'r')as file):
                data = json.load(file)
                for card in data["cards"]:
                    if card["card_number"] == self.card_number_inserted and card["pin"] == self.entered_pin:
                        card["balance"] += amount
            with(open("data.json", 'w')as file) :
                json.dump(data,file,indent=4)
            self.deposited_amount += amount
            print(f"Deposited: ${amount:.2f}")
        else:
            print("Invalid deposit amount. Please try again.")
        self.menu()
        
    def withdraw(self):
        amount = float(input("Enter the amount to withdraw: "))
        if 0 < amount <= self.balance:
            self.withdrawn_amount += amount
            self.balance -= amount
            print(f"Withdrew: ${amount:.2f}")
        else:
            print("Invalid withdrawal amount or insufficient funds. Please try again.")
        self.menu()

    def check_balance(self):
        print(f"Current Balance: ${self.balance:.2f}")
        self.menu()
    
    def transfer(self):
        amount = float(input("Enter the amount to transfer:"))
        if amount > 0 and amount <= self.balance:
            self.balance -=amount;
            print(f"Transferred: ${amount:.2f}")
        else:
            print("Invalid transfer amount or insufficient funds. Please try again.")
        self.menu()


    def save_pin(self):
        if 1000<=self.pin<=9999:
            with(open("data.json", 'r')as file):
                data = json.load(file)
            for card in data["cards"]:
                if card["card_number"] == self.card_number:
                    card["pin"] = self.pin
            with(open("data.json", 'w')as file):
                json.dump(data, file, indent=4)
            return True
        else:
            return False


    def change_pin(self):
        if(self.pin_authentication()):
            new_pin = int(input("enter new pin"))
            with(open("data.json",'r')as file):
                data = json.load(file)
            for card in data["cards"]:
                if(card["card_number"] == self.card_number_inserted):
                    card["pin"] = new_pin
            with(open("data.json",'w')as file):
                json.dump(data,file,indent = 4)
            print("pin changed successfully")
            
            
            
            






        

        
    
```
